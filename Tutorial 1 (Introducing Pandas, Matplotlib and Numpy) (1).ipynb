Use of .apply function

.apply() function is a pandas way to perform iterations on rows/columns. It takes advantage of vectorizes techniques and speeds up execution of simple and complex operations by many times.

```python
#The following example passes a function and checks the value of each element in series and returns low, normal 
or High accordingly.
def life_range(num): 
  
    if num < 0: 
        return "Low"
  
    elif num >= 0 and num < 1: 
        return "Normal"
  
    else: 
        return "High"
```

```python
#take an attribute 
Healthy_life = df["Healthy life expectancy"]
```

```python
Healthy_life
```

```python
# passing function to apply and storing returned series in range 
range = Healthy_life.apply(life_range)
```

```python
range
```

```python
df['Healthy Life Category'] = range #create new column in the dataset
```

```python
df
```

```python
df['Healthy Life Category'].value_counts() #Checks the ditribution of categorical columns
```

Data Visualization with Matplotlib:

Making informative visualizations (sometimes called plots) is one of the most important tasks in data analysis. It may be a part of the exploratory processâ€”for example,to help identify outliers or needed data transformations, or as a way of generating ideas for models. For others, building an interactive visualization for the web may be the end goal. Python has many add-on libraries for making static or dynamic visualizations, but Iâ€™ll be mainly focused on matplotlib and libraries that build on top of it.

```python
#simple histogram
n, bins, patches = plt.hist(x = df["Healthy life expectancy"],bins =6)
plt.setp(patches[0], 'facecolor', 'g')
plt.setp(patches[1], 'facecolor', 'r')

plt.xlabel('Life Expectancy')
plt.ylabel('Frequency')
plt.title("Histogram of Healthy Life Expectancy")
```

Numpy functions:

NumPy, short for Numerical Python, is one of the most important foundational packages for numerical computing and manipulation of data stored in arrays.

```python
a = np.array([9,10,17,3]) # One-dimensional array

b = np.array([[3, 5, 2, 4], 
              [7, 6, 8, 8],
              [1, 6, 7, 7]])  # Two-dimensional array

c = np.array([[8, 1, 5, 9], 
              [8, 9, 4, 3],
              [0, 3, 5, 0]])  # Two-dimensional array
```

Each array has attributes ndim (the number of dimensions), shape (the size of each dimension), and size (the total size of the array):

```python
print("Dimenssion of array a is : ", a.ndim)
print("Shape of array a is :", a.shape)
print("Size of array a is = ", a.size)
```

```python
print("Dimenssion of array b is : ", b.ndim)
print("Shape of array b is :", b.shape)
print("Size of array b is 3 * 4 = ", b.size)
```

```python
#multiplying a constant with array
scalar =  c*3
print(scalar)
```

```python
#adding b and c because they have same dimensions
add = b + c
print(add)
```

```python
# element wise multipication
multiply = b*c 
print(multiply)
```

```python
#taking square of every element in c
square = b**2
print(square)
```

```python
#are elements of b > than c
logical= b > c
print(logical)
```
